```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

sinasc = pd.read_csv('./sinasc_RO_2019.csv')
sinasc.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[(sinasc['APGAR5'] <= 10) & (sinasc['APGAR5'] >= 8), 'APGAR5_CAT'] = 'normal'
sinasc.loc[(sinasc['APGAR5'] <= 7) & (sinasc['APGAR5'] >= 6), 'APGAR5_CAT'] = 'asfixia leve'
sinasc.loc[(sinasc['APGAR5'] <= 5) & (sinasc['APGAR5'] >= 4), 'APGAR5_CAT'] = 'asfixia moderada'
sinasc.loc[sinasc['APGAR5'] <= 3, 'APGAR5_CAT'] = 'asfixia grave'
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
ax = sns.catplot(x='GESTACAO', y='APGAR5'
            , hue='SEXO'
            , kind='bar'
            , legend_out = True
            , data = sinasc
           )
           
```

```python
ax.set_xticklabels(rotation=20)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
freq = pd.crosstab(sinasc['GESTACAO'], sinasc['APGAR5'])
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq.sum(axis=1)
```

```python
freq_pclin = freq.div(freq.sum(axis=1), axis = 0)
freq_pclin
```

```python
freq_pclin.sum(axis=1)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
perc = freq_pclin.plot.bar(rot = 45)
perc
```

O conjunto de dados corrobora com a afirmação dos médicos, já que os maiores escores de APGAR5 se concentram, principalmente, em gestações de mais de 37 semanas. Além disso, os piores escores de APGAR5 se concentram em gestações com menos de 37 semanas.
